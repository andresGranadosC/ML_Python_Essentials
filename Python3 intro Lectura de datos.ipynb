# Lectura de datos con Astropy

https://www.kaggle.com/mrisdal/open-exoplanet-catalogue/version/2#

```python
filename = 'Datasets/oec.csv'
```

```python
from astropy.table import Table
```

```python
oec = Table.read(filename)
```

```python
oec
```

# Seleccionar columnas

```python
oec.mask
```

# Filtrar datos v√°lidos con Astropy

```python
oec['PlanetaryMassJpt'][:5]
```

```python
planetary_valid_data = oec['PlanetaryMassJpt'].mask == False
```

```python
planetary_valid_table = oec[planetary_valid_data]
```

```python
planetary_valid_table
```

```python
oec.columns
```

```python
oec['PlanetIdentifier'][:5]
```

```python
oec[['PlanetIdentifier', 'PlanetaryMassJpt']][:5]
```
